```python
import requests
import csv
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://wuzzuf.net/search/jobs/?a=navbl&q=data%20science%20jobs&start=1'
res=requests.get(url)
res.text
```

```python
soup=BeautifulSoup(res.text,'lxml')
soup
```

```python
tit=[]
for i in soup.find_all('div',attrs={'class':'css-laomuu'}):
  tit.append(i.find('h2',attrs={'class':'css-m604qf'}).find('a',attrs={'target':'_blank','rel':'noreferrer','class':'css-o171kl'}).get_text('href').strip())
tit
```

```python
domain='https://wuzzuf.net/'
```

```python
link=[]
for i in soup.find_all('div',attrs={'class':'css-laomuu'}):
  link.append(i.find('h2',attrs={'class':'css-m604qf'})
              .find_all('a',attrs={'target':'_blank','rel':'noreferrer','class':'css-o171kl'})[0]
              .get('href'))
link
```

```python
loc=[]
for i in soup.find_all('div',attrs={'class':'css-d7j1kk'}):
  loc.append(i.find_all('span',attrs={'class':'css-5wys0k'})[0].get_text().strip())
loc
```

```python
cmp=[]
for i in soup.find_all('div',attrs={'class':'css-d7j1kk'}):
  cmp.append(i.find('a').get_text().strip())
cmp
```

```python
sk=[]
for i in soup.find_all('div',attrs={'class':'css-1gatmva e1v1l3u10'}):
  sk.append(i.find('div',attrs={'class':'css-y4udm8'}).get_text().split('.')[0])
sk    
```

```python
with open('wuzzuf.csv','w',encoding='utf-8',newline='')as f:
  writer=csv.writer(f)
  writer.writerow(['Title_job','link','Location','company_name','skills'])
  for i in range(len(tit)):
   writer.writerow([tit[i],domain+link[i],loc[i],cmp[i],sk[i]])
```
